```python
import pickle
import numpy as np
```

```python
hands = {'グー':0, 'チョキ':1, 'パー':2}
results = ['あいこ', '負け', '勝ち']

judge = lambda a, b: (a - b + 3) % 3
```

```python
import random
random_hand = lambda : random.randint(0, 2)
x_items = []
y_items = []
for i in range(3000):
    a = random_hand()
    b = random_hand()
    result = judge(a, b)
    x_items.append([a, b])
    y_items.append(result)

print(x_items)
print(y_items)
```

```python
x_train = x_items[0:2000]
y_train = y_items[0:2000]
x_test = x_items[2000:]
y_test = y_items[2000:]

items = [[x_train, y_train],
        [x_test, y_test]]
```

```python
with open("./data/janken-data.pkl", "wb") as fp:
    pickle.dump(items, fp)

print("データ作成完了")
```
