```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../../images/ch3-flower.jpg")
img = cv2.resize(img, (300, 169))
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# img = cv2.GaussianBlur(img, (sx,ay), sigma_x)　← 平滑化、画像のぼかし処理
gray = cv2.GaussianBlur(gray, (7,7), 0)
# print(gray)
# ret, img = cv2.threshold(img, thresh, maxval, type)　← 二値化
im2 = cv2.threshold(gray, 140, 240, cv2.THRESH_BINARY_INV)
# print(im2)
im3 = im2[1]
# print(im3)
```

```python
plt.subplot(1,2,1)
plt.imshow(im3, cmap="gray")
```

```python
# contours, hierarchy = cv2.findContours(image, mode, method) ← 輪郭油出
cnts = cv2.findContours(im3, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]
# print(cnts)
```

```python
for pt in cnts:
    x, y, w, h = cv2.boundingRect(pt)
    if w < 30 or w > 200: continue
    print(x,y,w,h)
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),2)
```

```python
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.savefig("../../images/ch3-find_contours.png", dpi=200)
plt.show()
```
