```python
import matplotlib.pyplot as plt
import cv2

# https://github.com/opencv/opencv.git
cascade_file = "../../data/opencv/haarcascades/haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)
```

```python
img = cv2.imread("../../images/ch3-girl.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
face_list = cascade.detectMultiScale(img_gray, minSize=(150,150))

if len(face_list) == 0:
    print("Fail")
    quit()
```

```python
for (x, y, w, h) in face_list:
    print("face position =", x, y, w, h)
    red = (0, 0, 255)
    cv2.rectangle(img, (x, y), (x+w, y+h), red, thickness=10)
```

```python
cv2.imwrite("../../images/ch3-face-detect.png", img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```
