```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("../../images/ch3-test2.jpg")
# resize
# img = cv2.resize(origin-img, (img-width, img-height))
im2 = cv2.resize(img, (700,500))
cv2.imwrite("../../images/ch3-test2-resize.jpg", im2)
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
# cut out
# img[y1:y2, x1:x2]
im3 = img[150:200, 150:300] # 400x250
im3 = cv2.resize(im3, (800, 350))
cv2.imwrite("../../images/ch3-test2-cutout.jpg", im3)
plt.imshow(cv2.cvtColor(im3, cv2.COLOR_BGR2RGB))

plt.show()
```
