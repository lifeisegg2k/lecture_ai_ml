```python
import cv2
import matplotlib.pyplot as plt
from mosaic import mosaic as mosaic
```

```python
img = cv2.imread("../../images/ch3-cat.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
mos = mosaic(img, (50, 50, 450, 450), 10)
```

```python
cv2.imwrite("../../images/ch3-cat-mosaic.png", mos)
plt.imshow(cv2.cvtColor(mos, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
mos2 = mosaic(img, (100, 100, 400, 400), 20)
```

```python
cv2.imwrite("../../images/ch3-cat-mosaic2.png", mos2)
plt.imshow(cv2.cvtColor(mos2, cv2.COLOR_BGR2RGB))

plt.show()
```
