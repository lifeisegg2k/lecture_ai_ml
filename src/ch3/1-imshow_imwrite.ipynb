```python
import urllib.request as req

url = "https://image.shutterstock.com/image-photo/autumn-forest-nature-vivid-morning-260nw-766886038.jpg"
req.urlretrieve(url, "../../images/ch3-test2.jpg")
```

```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("../../images/ch3-test2.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
img = cv2.imread("../../images/ch3-test2.jpg")
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
img = cv2.imread("../../images/ch3-test2.jpg")

cv2.imwrite("../../images/ch3-test2-copy.jpg", img)

img = cv2.imread("../../images/ch3-test2-copy.jpg")
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```
