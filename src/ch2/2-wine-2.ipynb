```python
import pandas as pd
import matplotlib.pyplot as plt
savepath = "../../data/winequality-white.csv"
wine = pd.read_csv(savepath, sep=";", encoding="utf-8")
wine
```

```python
count_data = wine.groupby("quality")["quality"].count()
count_data
```

```python
count_data.plot()
plt.savefig("../../images/wine-count-plt.png")
plt.show
```
